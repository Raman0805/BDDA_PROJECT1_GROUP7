18. Airbags: The number of airbags in the car.

```python
# A function to model and also to calculate MAE
def get_mae(max_depth, X_train, val_X, y_train, val_y):
  # 3.1 Instantiate Decision tree regressor
    model = DecisionTreeRegressor(max_depth=max_depth)
    # 3.2 Train model
    model.fit(X_train, y_train)
    # 3.3 Make predictions
    preds_val = model.predict(val_X)
    # 3.4 Get error
    mae = mean_absolute_error(val_y, preds_val) #compare predicted val with the actual val #mean of absolute difference
    # 3.5 Return error
    return(mae)
    #get_mae, models a decision tree regressor with a specified maximum depth and calculates the Mean Absolute Error (MAE) for predictions on a validation set.
```

```python
# Compare MAE with differing values of max_depth:
#  Get

for max_depth in [2, 10, 20]:
    my_mae = get_mae(max_depth, val_X, X_test, val_y , y_test)
    print(f"Max depth: {max_depth} \t\t Mean Absolute Error: {my_mae}")

    #These results suggest that as the max_depth of the decision tree model increases, the MAE tends to decrease, indicating improved model performance on the test data. However, it's important to note that overly deep trees may lead to overfitting, where the model captures noise in the training data and performs poorly on new, unseen data.
```
